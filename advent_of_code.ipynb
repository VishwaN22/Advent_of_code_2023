Day 1 Puzzle 1

```python
with open ('input.txt','r') as file:

   
    line = file.readline().strip()
    sum =0
    while line:
        first=None
        last = None
        for char in line:
            if char.isdigit():
                if first is None:
                    first = char
                last = char

        callibration = int(first + last)
        sum+=callibration

        line = file.readline().strip()

    print(sum)

    file.close()
```

Day 1 Puzzle 2

```python
substrings =['one','two','three','four','five','six','seven','eight','nine']
with open ('input.txt','r') as file:

   
    line = file.readline().strip()
    sum =0
    while line:
        digits =[]
        for index,char in enumerate(line):
            if char.isdigit():
                digits.append(char)

            for i,substring in enumerate(substrings):
               if line[index:].startswith(substring):
                   digits.append(str(i + 1))

        callibration = int(digits[0] + digits[-1])
        sum+=callibration

        line = file.readline().strip()

    print(sum)

    file.close()
```
